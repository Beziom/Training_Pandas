```python
import pandas as pd
```

### 1.1.1. <a id='toc1_1_1_'></a>[Multindexing](#toc0_)

```python
"Importing data from Excel and setting Index from 0,1,2,3,4 to names"
df = pd.read_excel(r"D:\Repozytorium\Kody_Python\Python_ksiazka_przyklady\excel\course_participants.xlsx")
df
```

```python
"Setting Multiindexing"
df_multi = df.reset_index().set_index(["kontynent", "kraj"])
df_multi = df_multi.sort_index()
df_multi
```

```python
"Sorting index by Europe only"
df_multi.loc["Europa",:]
```

```python
df_multi.loc[("Europa", "Włochy")]
```

```python
"Reseting index to zero"
df_multi.reset_index(level=0)
```

### 1.1.2. <a id='toc1_1_2_'></a>[Data setting by position](#toc0_)

```python
"Creating a copy of df"
df2 = df.copy().set_index("numer")
df2
```

```python
"Setting column value for index 1000 to John in column imię"
df2.loc[1000,"imię"] = "JOHN"
df2
```

```python
"Setting"
df2.loc[[1000,1001],"ocena"] = [3,4]
df2
```

### 1.1.3. <a id='toc1_1_3_'></a>[Logic Indexing](#toc0_)

```python
"Logic indexing"
tf = (df2["wiek"]<20) | (df2["kraj"] == "USA") # Dataframe filtering wiek less than 20 or kraj = USA
df2.loc[tf,"imię"] = "xxx" # Setting all values in imię for "xxx"
df2
```

```python
rainfall[rainfall<400] = 0
rainfall
```

### 1.1.6. <a id='toc1_1_6_'></a>[Missing Data](#toc0_)
#### 1.1.6.1. <a id='toc1_1_6_1_'></a>[dropna(), fillna()](#toc0_)

```python
"Setting None/NaN values for new copy and displaying it"
df2 = df.copy().set_index("numer")
df2.loc[:,"Missing"] = [1,2,3,4]
df2.loc[1000,"ocena"] = None
df2.loc[1003,:] = None
df2
```

```python
"Delete all with missing data"
df2.dropna()
```

```python
"Delete only if all data are missing"
#dropna()
df2.dropna(how="all")
```

```python
"Filling out Nan values for mean from all column 'ocena'"
#fillna()
df2.fillna({"ocena":df2["ocena"].mean()})
```

### 1.1.7. <a id='toc1_1_7_'></a>[Duplikated Data](#toc0_)
#### 1.1.7.1. <a id='toc1_1_7_1_'></a>[is_unique, drop_duplicates(), unique(), duplicated()](#toc0_)

```python
"All duplikated value will be set as the first from row"
df.drop_duplicates(["kraj", "kontynent"]).set_index("numer")
```

```python
"Checking if there are duplicated datas in FataFrame"
df["kraj"].is_unique
```

```python
"Checking what unique datas are in DataFrame"
df["kraj"].unique()
```

```python
"Sprawdzanie kiedy występują duplikaty (without first attempt)"
df["kraj"].duplicated()
```

```python
"Checking rows which are duplicated "
df.loc[df["kraj"].duplicated(keep=False),:]
```
